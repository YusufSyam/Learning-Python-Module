```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data= pd.read_csv('dataset_popular_programming_languanges_cleaned.csv')
data
```

```python
data= data[data['Languange count']>data['Languange count'].mean()]
data
```

```python
slices= np.array([])
slc_labels= np.array([])

rand= np.random.randint(0,len(data)-1, 5)

for i in rand:
    slices= np.append(slices,data.iloc[i,1])
    slc_labels= np.append(slc_labels,data.iloc[i,0])

print(slices), print(slc_labels)
```

```python
# Menentukan bagian yang akan diexplode (yaitu bagian yang paling kecil)
explode= np.array([])
exploded_distance= 0.1

for i in range(len(slices)):
    if(np.where(slices==slices.min())[0]==[i]): explode= np.append(explode,exploded_distance)
    else: explode= np.append(explode, 0)

explode
```

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(8,8))
plt.pie(slices, labels=slc_labels, explode=explode, autopct='%1.1f%%')

plt.title('Survei seberapa banyak suatu bahasa pemrograman diketahui oleh orang-orang')

plt.tight_layout()
plt.show()
```
