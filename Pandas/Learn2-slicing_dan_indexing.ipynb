# Slicing dan indexing

```python
# Mengambil kolom
data_titanic['age']
# Atau
data_titanic.age
# Tidak bisa data_titanic[1]
```

```python
# Mengambil kolom tapi sebagai data frame
data_titanic[['age']]
```

```python
# Mengambil 2 kolom
data_titanic[['age','sex']]
```

```python
# Membandingkan apakah 2 kolom sama
data_titanic['age'].equals(data_titanic.age)
```

```python
data_titanic[['age']].equals(data_titanic.age)
```

```python
# Indexing dengan angka/index
# Dengan iloc[baris], atau iloc[baris,kolom]

# Mengambil baris ke lima dari data_titanic
data_titanic.iloc[5]
# Pakai loc[] jika ingin mengakses indeks berupa string/label
data_titanic.loc[3,'age'], data_titanic.iloc[3,3]
```

```python
# Mengambil semua baris dengan kolom 1 sampai 4
data_titanic.iloc[:,1:4]
```

```python
# Slicing dengan index negatif, misal dari 10 data terakhir sampai 3 data terakhir, dengan 2 kolom paling belakang
data_titanic.iloc[-10:-2,-2:]
# Tidak bisa begini data_titanic.iloc[-10:-20,-2:]
# Tidak bisa beginidata_titanic.iloc[-10:-2,-2:0]
```
