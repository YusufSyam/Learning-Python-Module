## Misc

#### Mendapatkan seberapa banyak para programmer Indonesia yang tau bahasa Java

```python
progLang= 'Java'
```

```python
# Bisa begini, pake variabel df
df[df['Country']=='Indonesia']['LanguageWorkedWith'].str.contains(progLang).sum()
```

Cara lain, tapi yang kita hitung bukan cuma Indonesia tapi semua

```python
df_knows_java= df_group['LanguageWorkedWith'].apply(lambda x: x.str.contains(progLang).sum())
df_knows_java
```

Mendapatkan persentase setiap negara yang tau java

```python
# Karena objek groupby tidak punya value_counts, maka kita perlu melakukan seperti ini
df_each_country_correspondents= df['Country'].value_counts()
df_each_country_correspondents
```

```python
df_pct_knows_java= pd.DataFrame(df_knows_java/df_each_country_correspondents) * 100
df_pct_knows_java= df_pct_knows_java.sort_index(0, ascending=False)
df_pct_knows_java= df_pct_knows_java.applymap(lambda x: '{:.2f} %'.format(x))
df_pct_knows_java.head()
```

Dan menggabungkan jumlah koresponden, jumlah yang tau java, dan persentase yang tau java

```python
new_df1= pd.concat([df_each_country_correspondents, df_knows_java, df_pct_knows_java], axis=1)
new_df1
```

```python
new_df1.rename(columns={'Country':'Correspondents', 'LanguageWorkedWith':'Knows Java', 0: 'Knows Java Percentage'}, inplace=True)

new_df1.fillna('0.00 %', inplace=True)
new_df1= new_df1.sort_values('Knows Java', ascending=False)
new_df1.head(50)
```
