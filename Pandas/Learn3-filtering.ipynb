```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('survey_results_public.csv')
df.head()
```

Filter data hanya dari negara India saja

```python
filt= df['Country']=='India'
df[filt]
```

Bisa juga seperti ini

```python
df[df['Country']=='India'].head()
```

Filter data hanya India **atau** Germany

```python
filt= (df['Country']=='India') | (df['Country']=='Germany')
df[filt]
```

Filter data orang bernegara India **dan** berumur di atas 20

```python
filt= (df['Country']=='India') & (df['Age']>20)
df[filt]
```
