```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_titanic= pd.read_csv('titanic_data_set.csv')
df_titanic.head(3)
```

# Histogram dari pandas

```python
# Misal ingin visualisasi histogram dari kolom age
df_titanic['age'].hist()

# Atau
df_titanic.hist()
```

```python
# Mempercantik histogram
# Arg bins pada hist(), akan menentukan seberapa besar batang pada histogram
# Arg alpha untuk opacity
# plot.hist() bukannya hist(), akan menghilangkan grid
df_titanic['age'].plot.hist(bins=(df_titanic['age'].nunique(dropna=True))) # Bins disesuaikan dengan jumlah age yang unik
```

```python
# Menggunakan library matplotlib
plt.hist(df_titanic['age'])
```

# Bar

```python
# Misal Data Frame baru
df_random1= pd.DataFrame(np.random.rand(5,5), columns=['A', 'B', 'C', 'D', 'E'])
df_random1, df_random1.plot.bar()
```

```python
df_random1.plot.bar(stacked=True, cmap='autumn_r')
```

# Scatter plot

```python
df_random2= pd.DataFrame(np.random.randint(100, size=(200,4)), columns=['A', 'B', 'C', 'D'])

df_random2.head(), df_random2.plot.scatter(x='A', y='B', c= 'C', cmap='Reds')
# argument x dan y pada plot.scatter merupakan 2 kolom yang akan dibandingkan terhadap argument kolom c
```

```python
# Bukan berdasarkan warna tapi berdasarkan besar kecil
df_random2.plot.scatter(x='A', y='B', s=df_random2['C'], c= 'C', cmap='Reds')
```

# Hex Bin

```python
# Pada dasarnya sama dengan scatter tapi bukannya titik tapi berbentuk hexagon
df_random2.plot.hexbin(x='A', y='B')
```

```python
# Arg gridsize untuk memperbesar hexagon
df_random2.plot.hexbin(x='A', y='B', gridsize=20, cmap='afmhot_r')
```

```python
# Membandingkan antara scatter dengan hexbin
plt.subplot(1,2,1)
plt.scatter(x=df_random2.A, y=df_random2.B, s=df_random2.C,c= df_random2.C, cmap='Reds')

plt.subplot(1,2,2)
plt.hexbin(x=df_random2.A, y=df_random2.B, gridsize=20, cmap='Reds')
```

# Plot Box

```python
df_random2.plot.box()
```

# Line

```python
df_random2.plot.line(figsize=(20,10))
```

```python
# Misal ingin hanya menampilkan line dari kolom A dan B
df_random2.plot.line(y=['A', 'B'], figsize=(20,5))
```

```python
# Misal hanya kolom c, dan garisnya dipertebal(arg lw)
df_random2.plot(y='C', lw=4, figsize=(20,5))
```
